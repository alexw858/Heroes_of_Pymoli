```python
# Most Popular Items by Purchase Count
# Groupby's on two columns(preserves those values), gets a count on the Age Group column (arbitrary), then sort_values
# popular = df.groupby(["Item ID","Item Name", ]).count()['Age Group'].sort_values(ascending=False)
# Convert GroupBy back into DataFrame, give the item purchase count column a new name, and display
# pd.DataFrame({'Count ':popular}).head()
# The problem is I can't get the sum of the item's prices this way

# I still need total purchase value for these items
# I don't want to group on "Price" column because I want to sum it
# I could group on Item Price, since each item is priced equally to itself, then multiply count by price

# THE STEPS
# create 3 groupby's on Item Id and item name
# One will do sum, one will do mean (for item price), one will do count
# then we create a new dataframe with these values inside groupby objects
# then sort descending

item_group = df.groupby(["Item ID", "Item Name"])
mostpop_items_count = item_group.count()["Price"]
mostpop_items_revenue = item_group.sum()["Price"]
mostpop_items_price = item_group.mean()["Price"]

most_popular_items_df = pd.DataFrame({
    "Item Count":mostpop_items_count, 
    "Item Revenue":mostpop_items_revenue, 
    "Item Price":mostpop_items_price
})
most_popular_items_df.sort_values(by="Item Count", ascending=False).head()
```
