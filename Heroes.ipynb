```python
#Age Demographics
# Age bins of 4 years: <10, 10-14. 15-19, etc.
bins = [0,10,14,18,22,26,30,34,38,42,46] #should be one more than group names (11 values here, 10 group names)
group_names = ["<10 Years", "10-14 Years", "14-18 Years", "18-22 Years", "22-26 Years", "26-30 Years", "30-34 Years", 
               "34-38 Years", "38-42 Years", "42-46 Years"]
```
